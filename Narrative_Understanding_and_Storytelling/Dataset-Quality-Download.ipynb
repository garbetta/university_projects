```python
import pandas as pd
import json
import csv
```

# Quality

```python
train_path = '/content/drive/MyDrive/NUANS/Project-MRC/DATASET/Quality/QuALITY.v1.0.1.htmlstripped.train'
val_path = '/content/drive/MyDrive/NUANS/Project-MRC/DATASET/Quality/QuALITY.v1.0.1.htmlstripped.dev'
test_path = '/content/drive/MyDrive/NUANS/Project-MRC/DATASET/Quality/QuALITY.v1.0.1.htmlstripped.test'
```

```python
def read_quality( path):
  json_data = []
  with open(path, 'r') as file:
      for line in file.readlines():
          json_obj = json.loads(line)
          json_data.append(json_obj)

  print(json_data[0])
  return json_data

train_quality = read_quality(train_path)
val_quality = read_quality(val_path)
test_quality = read_quality(test_path)
```

```python
#RETURN QUALITY DATA AS PANDAS DATAFRAME

def dataframe_quality(data_quality, flag):
  path_dest_data = '/content/drive/MyDrive/NUANS/Project-MRC/DATASET/Quality/'+flag+'.csv'
  mappa = {1: 'A', 2:'B', 3:'C', 4:'D'}
  with open(path_dest_data, 'w') as f:
    writer = csv.writer(f)
    if flag != 'test':
      headers=['question', "story", "answer_options", 'A', 'B', 'C', 'D', 'answer', 'label']
    else:
      headers=['question', "story", "answer_options", 'A', 'B', 'C', 'D']

    writer.writerow(headers)

    for i, elem in enumerate(data_quality):

        article = elem['article'].replace('\n', '')

        for j, question in enumerate(elem['questions']):
          que = question['question']

          opts = question['options']
          ans_opts = {'A':opts[0], 'B':opts[1], 'C':opts[2], 'D':opts[3]}

          if flag != 'test':
            label_num = question['gold_label']
            answer = mappa[label_num]
            label_num = label_num-1

            writer.writerow([que, article, ans_opts, opts[0], opts[1], opts[2], opts[3], answer, label_num])#
          else:
            writer.writerow([que, article, ans_opts, opts[0], opts[1], opts[2], opts[3]])
```

```python
dataframe_quality(train_quality, 'train')
dataframe_quality(val_quality, 'val')
dataframe_quality(test_quality, 'test')

data_train_qua = pd.readcsv('/content/drive/MyDrive/NUANS/Project-MRC/DATASET/Quality/train.csv')
data_val_qua = pd.readcsv('/content/drive/MyDrive/NUANS/Project-MRC/DATASET/Quality/val.csv')
data_test_qua = pd.readcsv('/content/drive/MyDrive/NUANS/Project-MRC/DATASET/Quality/test.csv')
```
